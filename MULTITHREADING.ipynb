```python
Q1). What is multithreading in python? Why is it used? Name the module used to handle threads in python.

ANS--Multithreading in Python refers to the ability of a program to execute multiple threads concurrently
within the same process. A thread is a lightweight subprocess that can execute independently, sharing the 
same resources as other threads within the process. Python's multithreading allows you to perform concurrent
tasks efficiently and take advantage of multi-core processors.

Multithreading is used to achieve concurrent execution and improve performance in certain scenarios. It allows 
a program to perform multiple tasks simultaneously, making better use of available resources and reducing the
overall execution time.

The threading module is used to handle threads. It provides classes and functions to create and manage threads
within a Python program. The threading module allows you to start, stop, and synchronize threads, and it also
provides mechanisms to manage shared resources and avoid conflicts between threads.
```

```python
Q2) Why threading module used? Write the use of the following functions
A) activeCount()
B) currentThread()
C) enumerate()

ANS--The threading module in Python is used to work with threads, which are lightweight sub-processes
that enable concurrent execution of tasks within a program. It allows you to create, start, stop, and 
synchronize threads, making it easier to implement concurrent programming in Python.

A) activeCount():
The activeCount() function returns the number of Thread objects that are currently alive (i.e., still running) 
in the Python program.

B) currentThread():
The currentThread() function returns the Thread object representing the currently executing thread.

C) enumerate():
The enumerate() function returns a list of all Thread objects that are currently alive (running) in the Python
program. It is useful for getting a list of all active threads for further processing or monitoring.

```

```python
3. Explain the following functions
A) run()
B) start()
 C)join()
D) isAlive()

ANS--A) run():
The run() method is not a standalone function but a method that you can override when you create 
a custom thread class by subclassing threading.Thread

B) start():
The start() method is used to begin the execution of the thread. It creates a new thread of execution, calls 
the run() method of the thread, and starts executing the code defined in the run() method concurrently with
the main thread.

C) join():
The join() method is used to wait for a thread to complete its execution. It blocks the calling thread
(usually the main thread) until the target thread (the thread on which join() is called) finishes its execution. 
This is helpful when you want to ensure that certain tasks are completed in order before the main thread proceeds.

D) isAlive():
The isAlive() method is used to check if a thread is currently running or active. It returns True if the thread 
has started and is still running, and False if the thread has completed its execution.
```

```python
4. rite a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes

ANS--
import threading

def print_squares(numbers):
    for num in numbers:
        print(f"Square of {num}: {num ** 2}")

def print_cubes(numbers):
    for num in numbers:
        print(f"Cube of {num}: {num ** 3}")

if __name__ == "__main__":
    numbers = [1, 2, 3, 4, 5]

    # Create two thread instances with target functions
    thread1 = threading.Thread(target=print_squares, args=(numbers,))
    thread2 = threading.Thread(target=print_cubes, args=(numbers,))

    # Start both threads
    thread1.start()
    thread2.start()

    # Wait for both threads to finish
    thread1.join()
    thread2.join()

    print("Both threads have completed their execution.")
```
